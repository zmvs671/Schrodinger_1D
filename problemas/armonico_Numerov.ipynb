Resolvemos numéricamente la ecuación de Schrödinger dependiente del tiempo en una dimensión para un potencial armónico:

$$
-\frac{\hbar^2}{2m} \frac{d^2\psi(x)}{dx^2} + V(x)\psi(x) = E\psi(x) 
$$

con un potencial de la forma:
$$ V(x) = \frac{1}{2}m\omega^2 x^2 $$

*Basado en [4]*


```python
import numpy as np
from matplotlib import pyplot as plt
import sys
sys.path.append("../metodos")
from libMetodos import numerov
```

```python
# Definicion del potencial armonico: V(x) = (1/2) * m * omega^2 * x^2
def V(x, m=1.0, omega=1.0):
    return 0.5 * m * omega**2 * x**2
```

```python
# Parametros
x_min = -6
x_max = 6
h = 0.001
E = 2.5  # Propuesta de valor de energía (n + 1/2)
```

```python
# Ejecutar el metodo de Numerov
x, psi = numerov(V, x_min, x_max, h, E)
```

```python
# Graficar
plt.figure(figsize=(10, 5))
plt.plot(x, psi, label=f"$E = {E}$")
plt.title("Solución de la ecuación de Schrodinger con potencial armónico")
plt.xlabel("x")
plt.ylabel("$\\psi(x)$")
plt.legend()
plt.grid(True)
plt.show()
```
